```python
import pandas as pd
df = pd.read_csv("Filtered_Companies_Capitalized.csv")  


input_fields = ["input_main_country_code", "input_main_country", "input_main_region", 
                "input_main_city", "input_main_postcode", "input_main_street", "input_main_street_number"]

company_fields = ["main_country_code", "main_country", "main_region", 
                  "main_city", "main_postcode", "main_street", "main_street_number"]

def check_match(row):
    for input_field, company_field in zip(input_fields, company_fields):
        input_value = str(row[input_field]).strip().lower() if pd.notna(row[input_field]) else ""
        company_value = str(row[company_field]).strip().lower() if pd.notna(row[company_field]) else ""
        if input_value != company_value:
            return "unmatched"
    return "matched"


df["match_status"] = df.apply(check_match, axis=1)


df.to_csv("Sample_Companies.csv", index=False)
```
